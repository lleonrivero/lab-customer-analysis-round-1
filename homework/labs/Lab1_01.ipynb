# 6. Perform the data cleaning operations mentioned so far in class

Filter out the data for customers who have an income of 0 or less.

```python
# Delete the column education and the number of open complaints from the dataframe. 
data = data.drop(['education', 'number_of_open_complaints'], axis=1)
data
```

```python
# Check for duplicate rows in the data and remove if any.

data = data.drop_duplicates()
data = data.reset_index(drop=True)
data

# Around 3.000 rows removed. As 'data' is the result of merging 'file1', 'file2' and 'file3', these records must have been present in several datasets.
```

```python
# Filter out the data for customers who have an income of 0 or less.
# Also trying to reset index through same function. It works!
data[data['income']>0].reset_index(drop=True)
```

```python
# Correct the values in the column customer lifetime value. 
# They are given as a percent, so multiply (--> divide) them by 100 and change dtype to numerical type. 

# Checking for NaN values in column customer_lifetime_value. 
data[data['customer_lifetime_value'].isna()==True].head(60)


# Trying to convert str into int or float, but not working. 
# df = pd.DataFrame(data)
# df['customer_lifetime_value'] = df['customer_lifetime_value'].astype(int)

```
